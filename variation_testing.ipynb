```python
#see complete_profiles and oned_profiles for development
import time
import complete_profiles as cmp
import numpy as np
from matplotlib.pyplot import Normalize as normalize
from matplotlib import pylab as plt
from matplotlib import cm
from matplotlib import colors as clr
  
#figure setup
viridis = cm.get_cmap('viridis', 20)

x_array = np.linspace(-4,4,512)

#calculation and plotting
file_path = './ad_prof/amp_1/sig_1'

for i in range(20):
    #print('passing step ',i)
    if i == 0:
        fname = file_path+'/HGB.out2.00100.athdf'
    else:
        fname = file_path+'/HGB.out2.000'+str(100-i)+'.athdf'
    betabar_list=cmp.oned_betabar_profile(fname)
    if max(betabar_list) > 6000:
        plt.plot(x_array,betabar_list,label = str(fname))

#color bar setup
plt.legend()
crange=np.linspace(80,100,21)
mappabl = cm.ScalarMappable(norm = clr.Normalize(vmin = 5,vmax=10))
#plt.colorbar(mappabl)
plt.ylabel('$\\overline{2\\rho/B^2}$')
plt.xlabel('x/H')
plt.title('$\\overline{\\beta}$ azimuthal average profile over last 20 orbits')    

```
