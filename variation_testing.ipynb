```python
#look for holes in bmag squared profile, specifically in step 87

import sys
import time
sys.path.append('~/athena-public-version/vis/python/')
#sys.path.append('~/.local/lib/python3.8/site-packages/')
sys.path.append('~/working')


import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
#IMPORT APPROPRIATE ATHINPUT FILE
#import athinput.hgb as athin
import athena_read


file_name = './ad_prof/amp_1/sig_1/HGB.out2.00087.athdf'

data = athena_read.athdf(file_name)
bx = data['Bcc1']
by = data['Bcc2']
bz = data['Bcc3']

bsquared = bx*bx + by*by + bz*bz

bsquared_min = np.min(bsquared,axis=0)

#plotting
plt.pcolormesh(bsquared_min,norm=mpl.colors.LogNorm(vmin=.0000001),shading = 'gouraud',cmap = 'RdBu_r')
plt.colorbar()
plt.xlabel('x cell')
plt.ylabel('y cell')
plt.title('$B^2$ vertical min map for Orbit 87')
```
