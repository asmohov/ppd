```python
from eprime_y import eprime_y
import numpy as np
from matplotlib import pylab as plt

#print('eprime is ',eprime_y('./ad_prof/const_am/HGB.out2.00040.athdf'))


eprime_tavg = []

for i in range(20,89):
    eprime_tavg.append(eprime_y('./ad_prof/const_am/HGB.out2.000'+str(i)+'.athdf'))
    #print(i)
    #print('current eprime is ',eprime_tavg[-1])
print('overall avg is ',np.average(eprime_tavg))
plt.plot(range(20,89),eprime_tavg)
plt.xlabel('period')
plt.ylabel('vol avgd eprime_y')
plt.show()

```
