```python
import sys
sys.path.append('~/athena-public-version/vis/python/')
#sys.path.append('~/.local/lib/python3.8/site-packages/')
sys.path.append('~/working')
import athena_read
from matplotlib import pylab as plt
import numpy as np
import matplotlib as mpl



def rho_rms_calculator(file_path,min_orbits = 20, max_orbits = 88):
    rms_arr = []
    arr = range(min_orbits,max_orbits+1)
    for i in arr:
        data = athena_read.athdf(file_path+'/HGB.out2.000'+str(i)+'.athdf')['rho']
        data = (1-data)**2
        rho_rms = np.sqrt(np.average(data))
        rms_arr.append(rho_rms)
    return rms_arr
def rho_min_calculator(file_path,min_orbits = 20, max_orbits = 88):
    min_arr = []
    arr = range(min_orbits,max_orbits+1)
    for i in arr:
        data = athena_read.athdf(file_path+'/HGB.out2.000'+str(i)+'.athdf')['rho']
        min_arr.append(np.min(data))
    return min_arr
def rho_max_calculator(file_path,min_orbits = 20, max_orbits = 88):
    max_arr = []
    arr = range(min_orbits,max_orbits+1)
    for i in arr:
        data = athena_read.athdf(file_path+'/HGB.out2.000'+str(i)+'.athdf')['rho']
        max_arr.append(np.max(data))
    return max_arr

arr = range(20,89)
arr2 = range(20,100)

#rms
arr_4x4 = rho_rms_calculator('./ad_prof/const_am')
arr_8x8 = rho_rms_calculator('./ad_prof/const_am_big/1point0',max_orbits = 99)
arr_point1 = rho_rms_calculator('./ad_prof/const_am_big/point1',max_orbits = 99)

#min
arr_4x4min = rho_min_calculator('./ad_prof/const_am')
arr_8x8min = rho_min_calculator('./ad_prof/const_am_big/1point0',max_orbits = 99)
arr_point1min = rho_min_calculator('./ad_prof/const_am_big/point1',max_orbits = 99)
#max
arr_4x4max = rho_max_calculator('./ad_prof/const_am')
arr_8x8max = rho_max_calculator('./ad_prof/const_am_big/1point0',max_orbits = 99)
arr_point1max = rho_rms_calculator('./ad_prof/const_am_big/point1',max_orbits = 99)

plt.plot(arr,arr_4x4min,label = '4x4 rho min', color = 'r')
plt.plot(arr2,arr_8x8min,label = '8x8 rho min', color = 'b')
plt.plot(arr2,arr_point1min,label = 'point1 rho min',color = 'g')


plt.plot(arr,arr_4x4max,label = '4x4 rho max', color = 'r')
plt.plot(arr2,arr_8x8max,label = '8x8 rho max', color = 'b')
plt.plot(arr2,arr_point1max,label = 'point1 rho max',color ='g')
plt.legend()
plt.show()

```
