```python
#Computational outputs from athena++ simulations comparison
#fast magnetosonic
fig1,axes = plt.subplots(3,3)
fig1.set_size_inches(15,10)
fig1.suptitle('Fast Magnetosonic 1-d Wave')

rho_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
velx_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
vely_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
velz_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
press_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
by_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')
bz_plot('./double_outputs/LinWave.block0.out2.00000.vtk','t=00')



rho_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
velx_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
vely_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
velz_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
press_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
by_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')
bz_plot('./double_outputs/LinWave.block0.out2.00010.vtk','t=10')

fig1.tight_layout()


#Analytic solution for Alfvenic waves

import numpy as np

#Defined variables (taken from input file)
gamma = (5/3)
k = np.pi*2
cs = 1
amp = .0001
dt = .01
t = dt*4000
B_0 = np.sqrt(3.25) #calculate from code
rho_0 = 1
theta = .9823 #RAD, calculate from source code

#from code
bx0 = 1.0
by0 = np.sqrt(2.0)
bz0 = 0.5

#Derived variables
#alfven velocity
v_a = np.sqrt((B_0**2)/(rho_0))
press_0 = 1/gamma


print('alfven velo = ',v_a)
#dispersion relation for alfvenic waves
omega_alf = np.sqrt((k**2)*(v_a**2)*((np.cos(theta)**2)))

#dispersion relation for fast
omega_slow_sq = .5*(k**2)*((v_a**2)+(cs**2)-np.sqrt((v_a**2+cs**2)**2-4*v_a**2*cs**2*(np.cos(theta))**2))
omega_slow = np.sqrt(omega_slow_sq)
#dispersion relation for slow
omega_fast_sq = .5*(k**2)*((v_a**2)+(cs**2)+np.sqrt((v_a**2+cs**2)**2-4*v_a**2*cs**2*(np.cos(theta))**2))
omega_fast = np.sqrt(omega_fast_sq)
#analytical equation for rho, taking the sin part

ws_fast = omega_fast/k
ws_slow = omega_slow/k
print('ws_fast is ',ws_fast)
print('ws_slow is ',ws_slow)
print('period is',np.pi*2/omega_fast)
#perturbation amplitude factors from athena paper

alpha_f = np.sqrt((cs**2-ws_slow**2)/(ws_fast**2-ws_slow**2))
C_ff = ws_fast*alpha_f
alpha_s = np.sqrt((ws_fast**2-cs**2)/(ws_fast**2-ws_slow**2))
C_ss = ws_slow*alpha_s
beta_z = bz0/np.sqrt(by0**2+bz0**2)
beta_y = by0/np.sqrt(by0**2+bz0**2)
A_f = cs*np.sqrt(rho_0)*alpha_f
A_s = cs*np.sqrt(rho_0)*alpha_s
#since sgn(bx) is constant and positive, Q_f from the paper is same as C_ff
#similarly, Q_s = C_ss
Q_f= C_ff
Q_s =C_ss


#perturbation eigenvector
rho_pert=amp*alpha_f
vx_pert = -amp*C_ff
vy_pert = amp*Q_s*beta_y
vz_pert = amp*Q_s*beta_z
press_pert = amp*rho_0*(cs**2)*alpha_f
by_pert = amp*A_s*beta_y
bz_pert = amp*A_s*beta_z

print('vy initial is ',vy_pert)

ana_rho = []
ana_vx = []
ana_vy = []
ana_vz = []
ana_press = []
ana_by = []
ana_bz = []

#second set of vectors for time comparison
ana_rho2 = []
ana_vx2 = []
ana_vy2 = []
ana_vz2 = []
ana_press2 = []
ana_by2 = []
ana_bz2 = []

#define analytical solution

def fast_rho(x,t):
    return np.sin(k*x+omega_fast*t)*rho_pert + rho_0
def fast_vx(x,t):
    return np.sin(k*x+omega_fast*t)*vx_pert
def fast_vy(x,t):
    return np.sin(k*x+omega_fast*t)*vy_pert
def fast_vz(x,t):
    return np.sin(k*x+omega_fast*t)*vz_pert
def fast_press(x,t):
    return np.sin(k*x+omega_fast*t)*press_pert+press_0
def fast_by(x,t):
    return np.sin(k*x+omega_fast*t)*by_pert + by0
def fast_bz(x,t):
    return np.sin(k*x+omega_fast*t)*bz_pert + bz0

#iterate analytical solution over x domain

lin_space = np.linspace(0,1,1000)
for x in lin_space:

    ana_rho.append(fast_rho(x,0*dt))
    ana_vx.append(fast_vx(x,0*dt))
    ana_vy.append(fast_vy(x,0*dt))
    ana_vz.append(fast_vz(x,0*dt))
    ana_press.append(fast_press(x,0*dt))
    ana_by.append(fast_by(x,0*dt))
    ana_bz.append(fast_bz(x,0*dt))
    #for time comparison, 2nd set
    ana_rho2.append(fast_rho(x,10*dt))
    ana_vx2.append(fast_vx(x,10*dt))
    ana_vy2.append(fast_vy(x,10*dt))
    ana_vz2.append(fast_vz(x,10*dt))
    ana_press2.append(fast_press(x,10*dt))
    ana_by2.append(fast_by(x,10*dt))
    ana_bz2.append(fast_bz(x,10*dt))
    
    
    
#plot analytical solutions
axes[0][0].plot(lin_space,ana_rho,'k--',label='analytical')
axes[0][0].plot(lin_space,ana_rho2,'--',label='analytical2')
axes[0][0].legend()

axes[0][1].plot(lin_space,ana_press,'k--',label='analytical')
axes[0][1].plot(lin_space,ana_press2,'--',label='analytical2')
axes[0][1].legend()

axes[1][0].plot(lin_space,ana_vx,'k--',label='analytical')
axes[1][0].plot(lin_space,ana_vx2,'--',label='analytical2')
axes[1][0].legend()

axes[1][1].plot(lin_space,ana_vy,'k--',label='analytical')
axes[1][1].plot(lin_space,ana_vy2,'--',label='analytical2')
axes[1][1].legend()

axes[1][2].plot(lin_space,ana_vz,'k--',label='analytical')
axes[1][2].plot(lin_space,ana_vz2,'--',label='analytical2')
axes[1][2].legend()

axes[2][1].plot(lin_space,ana_by,'k--',label='analytical')
axes[2][1].plot(lin_space,ana_by2,'--',label='analytical2')
axes[2][1].legend()

axes[2][2].plot(lin_space,ana_bz,'k--',label='analytical')
axes[2][2].plot(lin_space,ana_bz2,'--',label='analytical2')
axes[2][2].legend()

fig1.show()
```

```python
#Computational outputs from athena++ simulations comparison
#slow, moving right
fig1,axes = plt.subplots(3,3)
fig1.set_size_inches(15,10)
fig1.suptitle('Slow Magnetosonic 1-D wave')

rho_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
velx_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
vely_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
velz_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
press_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
by_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')
bz_plot('./double_outputs_slow/LinWave.block0.out2.00000.vtk','t=00')



rho_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
velx_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
vely_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
velz_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
press_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
by_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')
bz_plot('./double_outputs_slow/LinWave.block0.out2.00010.vtk','t=10')

fig1.tight_layout()


#Analytic solution for slow magnetosonic waves

import numpy as np

#Defined variables (taken from input file)
gamma = (5/3)
k = np.pi*2
cs = 1
amp = .0001
dt = .01
t = dt*4000
B_0 = np.sqrt(3.25) #calculate from code
rho_0 = 1
theta = .9823 #RAD, calculate from source code

#from code
bx0 = 1.0
by0 = np.sqrt(2.0)
bz0 = 0.5

#Derived variables
#alfven velocity
v_a = np.sqrt((B_0**2)/(rho_0))
press_0 = 1/gamma
print('press_0 is ',press_0)

print('alfven velo = ',v_a)
#dispersion relation for alfvenic waves
omega_alf = np.sqrt((k**2)*(v_a**2)*((np.cos(theta)**2)))

#dispersion relation for fast
omega_slow_sq = .5*(k**2)*((v_a**2)+(cs**2)-np.sqrt((v_a**2+cs**2)**2-4*v_a**2*cs**2*(np.cos(theta))**2))
omega_slow = np.sqrt(omega_slow_sq)
#dispersion relation for slow
omega_fast_sq = .5*(k**2)*((v_a**2)+(cs**2)+np.sqrt((v_a**2+cs**2)**2-4*v_a**2*cs**2*(np.cos(theta))**2))
omega_fast = np.sqrt(omega_fast_sq)
#analytical equation for rho, taking the sin part

ws_fast = omega_fast/k
ws_slow = omega_slow/k
print('ws_fast is ',ws_fast)
print('ws_slow is ',ws_slow)
print('period is',np.pi*2/omega_slow)
#perturbation amplitude factors from athena paper

alpha_f = np.sqrt((cs**2-ws_slow**2)/(ws_fast**2-ws_slow**2))
C_ff = ws_fast*alpha_f
alpha_s = np.sqrt((ws_fast**2-cs**2)/(ws_fast**2-ws_slow**2))
C_ss = ws_slow*alpha_s
beta_z = bz0/np.sqrt(by0**2+bz0**2)
beta_y = by0/np.sqrt(by0**2+bz0**2)
A_f = cs*np.sqrt(rho_0)*alpha_f
A_s = cs*np.sqrt(rho_0)*alpha_s
#since sgn(bx) is constant and positive, Q_f from the paper is same as C_ff
#similarly, Q_s = C_ss
Q_f= C_ff
Q_s =C_ss

print('alpha_f is ',alpha_f)
print('alpha_s is ',alpha_s)

print('two times alpha f squared is ',2*(alpha_f**2))
print('two times alpha s squared is ',2*(alpha_s**2))


#perturbation eigenvector, for slow magnetosonic, moving right
rho_pert=amp*alpha_s*rho_0
vx_pert = amp*C_ss
vy_pert = amp*Q_f*beta_y
vz_pert = amp*Q_f*beta_z
press_pert = amp*rho_0*(cs**2)*alpha_s
by_pert = -amp*A_f*beta_y
bz_pert = -amp*A_f*beta_z



ana_rho = []
ana_vx = []
ana_vy = []
ana_vz = []
ana_press = []
ana_by = []
ana_bz = []

#second set of vectors for time comparison
ana_rho2 = []
ana_vx2 = []
ana_vy2 = []
ana_vz2 = []
ana_press2 = []
ana_by2 = []
ana_bz2 = []

#define analytical solution

def slow_rho(x,t):
    return np.sin(k*x-omega_slow*t)*rho_pert + rho_0
def slow_vx(x,t):
    return np.sin(k*x-omega_slow*t)*vx_pert
def slow_vy(x,t):
    return np.sin(k*x-omega_slow*t)*vy_pert
def slow_vz(x,t):
    return np.sin(k*x-omega_slow*t)*vz_pert
def slow_press(x,t):
    return (np.sin(k*x-omega_slow*t)*press_pert) + press_0
def slow_by(x,t):
    return np.sin(k*x-omega_slow*t)*by_pert + by0
def slow_bz(x,t):
    return np.sin(k*x-omega_slow*t)*bz_pert + bz0

#iterate analytical solution over x domain

lin_space = np.linspace(0,1,1000)
for x in lin_space:

    ana_rho.append(slow_rho(x,0*dt))
    ana_vx.append(slow_vx(x,0*dt))
    ana_vy.append(slow_vy(x,0*dt))
    ana_vz.append(slow_vz(x,0*dt))
    ana_press.append(slow_press(x,0*dt))
    ana_by.append(slow_by(x,0*dt))
    ana_bz.append(slow_bz(x,0*dt))
    #for time comparison, 2nd set
    ana_rho2.append(slow_rho(x,10*dt))
    ana_vx2.append(slow_vx(x,10*dt))
    ana_vy2.append(slow_vy(x,10*dt))
    ana_vz2.append(slow_vz(x,10*dt))
    ana_press2.append(slow_press(x,10*dt))
    ana_by2.append(slow_by(x,10*dt))
    ana_bz2.append(slow_bz(x,10*dt))
    
    
    
#plot analytical solutions
axes[0][0].plot(lin_space,ana_rho,'k--',label='analytical')
axes[0][0].plot(lin_space,ana_rho2,'--',label='analytical2')
axes[0][0].legend()

axes[0][1].plot(lin_space,ana_press,'k--',label='analytical')
axes[0][1].plot(lin_space,ana_press2,'--',label='analytical2')
axes[0][1].legend()

axes[1][0].plot(lin_space,ana_vx,'k--',label='analytical')
axes[1][0].plot(lin_space,ana_vx2,'--',label='analytical2')
axes[1][0].legend()

axes[1][1].plot(lin_space,ana_vy,'k--',label='analytical')
axes[1][1].plot(lin_space,ana_vy2,'--',label='analytical2')
axes[1][1].legend()

axes[1][2].plot(lin_space,ana_vz,'k--',label='analytical')
axes[1][2].plot(lin_space,ana_vz2,'--',label='analytical2')
axes[1][2].legend()

axes[2][1].plot(lin_space,ana_by,'k--',label='analytical')
axes[2][1].plot(lin_space,ana_by2,'--',label='analytical2')
axes[2][1].legend()

axes[2][2].plot(lin_space,ana_bz,'k--',label='analytical')
axes[2][2].plot(lin_space,ana_bz2,'--',label='analytical2')
axes[2][2].legend()

fig1.show()
print('vy initial is ',vy_pert)
```
