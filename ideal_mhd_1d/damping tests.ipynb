```python
import numpy as np
from matplotlib import pylab as plt

x_arr = np.linspace(0,4*np.pi,1000)

#define damping rate
d = .125
w = .15*np.pi
k=1
def func1(x,t):
    return np.exp(-d*t)*np.sin(k*x-w*t)

t_arr = np.linspace(0,20,7)



for t in t_arr:
    
    y_arr = []
    for x in x_arr:        
        y_arr.append(func1(x,t))    
    plt.plot(x_arr,y_arr,label = str(t))


plt.rcParams["figure.figsize"] = (10,10)
plt.legend()

lim_x=[]
lim_y=[]
lim = 0

#parametric bounding function
for t in t_arr:
    #add the first max location to the x array
    lim_x.append((np.pi/2)+w*t)
    lim_y.append(np.exp(-d*t))
#plt.plot (lim_x,lim_y)
    
    
#analytic solution to the bounding function, independent of t
ana_lim = []
ana_lim2 = []
ana_lim3 = []
ana_lim4 = []
ana_lim5 = []
ana_lim6 = []
for x in x_arr:
    num = k*x-np.pi/2
    num2 = k*x+3*np.pi/2
    num3 = k*x-5*np.pi/2
    num4 = k*x+np.pi/2
    num5 = k*x+5*np.pi/2
    num6 = k*x-np.pi*3/2
    ana_lim.append(np.exp(num*-d/w))
    ana_lim2.append(np.exp(num2*-d/w))
    ana_lim3.append(np.exp(num3*-d/w))
    ana_lim4.append(-np.exp(num4*-d/w))
    ana_lim5.append(-np.exp(num5*-d/w))
    ana_lim6.append(-np.exp(num6*-d/w))
plt.plot(x_arr,ana_lim,'k:')
plt.plot(x_arr,ana_lim2,'k:')
plt.plot(x_arr,ana_lim3,'k:')
plt.plot(x_arr,ana_lim4,'k:')
plt.plot(x_arr,ana_lim5,'k:')
plt.plot(x_arr,ana_lim6,'k:')
#plt.plot(lim_x,lim_y)

#here the argument of the limit function exponential is d/w where w is the time frequency of the wave
plt.xlim([0,4*np.pi])
plt.ylim([-1,1])
plt.show()
#now test with arbitrary time steps to make sure this holds for all tiem
plt.plot(x_arr,ana_lim,'k:')
plt.plot(x_arr,ana_lim2,'k:')
plt.plot(x_arr,ana_lim3,'k:')
plt.plot(x_arr,ana_lim4,'k:')
plt.plot(x_arr,ana_lim5,'k:')
plt.plot(x_arr,ana_lim6,'k:')
t_arr = np.linspace(0,20,151)
y_arr = []

for t in t_arr:
    
    y_arr = []
    for x in x_arr:        
        y_arr.append(func1(x,t))    
    plt.plot(x_arr,y_arr,label = str(t))

plt.rcParams["figure.figsize"] = (10,10)
#plt.legend()
plt.xlim([0,4*np.pi])
plt.ylim([-1,1])
plt.show()
```
