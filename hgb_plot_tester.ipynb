```python
import sys
sys.path.append('~/athena-public-version/vis/python/')
#sys.path.append('~/.local/lib/python3.8/site-packages/')
sys.path.append('~/working')

import numpy as np
import matplotlib.pyplot as plt
import athena_read

data = athena_read.hst('./ad_prof/const_am/HGB.hst')



#set x values as time
x_vals=((data['time']))

#to match HGB, divide by 2 pi, because of difference in code units: 1/omega vs period
x_vals_adj = x_vals/(2*np.pi)

print(data)
print('length of x value array ',len(x_vals_adj))
volume =16
sat_1_KE = data['1-KE'][2000:]
sat_2_KE = data['2-KE'][2000:]
sat_3_KE = data['3-KE'][2000:]
sat_1_ME = data['1-ME'][2000:]
sat_2_ME = data['2-ME'][2000:]
sat_3_ME = data['3-ME'][2000:]
sat_mag_stress = data['-BxBy'][2000:]
sat_rey_stress = data['dVxVy'][2000:]


print(sat_1_KE)

#energy projections --------------------
KE_1_avg = (np.average(sat_1_KE))/volume
KE_2_avg = (np.average(sat_2_KE))/volume
KE_3_avg = (np.average(sat_3_KE))/volume
ME_1_avg = (np.average(sat_1_ME))/volume
ME_2_avg = (np.average(sat_2_ME))/volume
ME_3_avg = (np.average(sat_3_ME))/volume
#stress parameters---------------------

alpha_rey = np.average(sat_rey_stress)/(volume)
alpha_mag = np.average(sat_mag_stress)/(volume)

print('1_KE_avg is',KE_1_avg)
print('2_KE_avg is',KE_2_avg)
print('3_KE_avg is',KE_3_avg)
print('1_ME_avg is',ME_1_avg)
print('2_ME_avg is',ME_2_avg)
print('3_ME_avg is',ME_3_avg)
print('alpha_rey is ',alpha_rey)
print('alpha_mag is ',alpha_mag)
print('alpha total is ',(alpha_mag+alpha_rey))
```
