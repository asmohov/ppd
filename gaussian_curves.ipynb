```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

viridis = cm.get_cmap('viridis', 101)

x_array = np.linspace(-4,4,100)

def gaussian(x,sigma):
    return np.exp(-(x**2)/(2*(sigma**2)))               
for sig in np.linspace(.25,10.25,101):
    y_array = []
    for equis in x_array:
        y_array.append(gaussian(equis,sig))

    if (sig-.25)%1== 0 :
        labl = 'sig='+str(sig)
        plt.plot(x_array,y_array,color=viridis(sig/5.25),label=labl)
    else:
        plt.plot(x_array,y_array,color=viridis(sig/5.25))
      
plt.legend(loc='upper right') 
plt.show()
```
